```python
import numpy as np
import matplotlib.pyplot as plt 

# generate some data
x = np.arange(0, 10, 0.2)
y = np.sin(x)

# plot it
f, (a0, a1) = plt.subplots(1, 2, gridspec_kw={'width_ratios': [3, 1]})
a0.plot(x, y)
a1.plot(y, x)

f.tight_layout()
f.savefig('grid_figure.pdf')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# generate some data
x = np.arange(0, 10, 0.2)
y = np.sin(x)

# plot it
fig = plt.figure(figsize=(8, 4))
ax0 = plt.subplot2grid((1, 3), (0, 0), colspan=1)
ax0.plot(x, y)
ax1 = plt.subplot2grid((1, 3), (0, 2))
ax1.plot(y, x)

plt.tight_layout()
plt.savefig('grid_figure.pdf')
```

```python
import matplotlib.pyplot as plt

# On-screen, things will be displayed at 80dpi regardless of what we set here
# This is effectively the dpi for the saved figure. We need to specify it,
# otherwise `savefig` will pick a default dpi based on your local configuration
dpi = 80

im_data = plt.imread('test_images/solidWhiteRight.jpg')
height, width, nbands = im_data.shape

# What size does the figure need to be in inches to fit the image?
figsize = width / float(dpi), height / float(dpi)

# Create a figure of the right size with one axes that takes up the full figure
fig = plt.figure(figsize=figsize)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(5, 5), dpi=80, sharex=True, sharey=True)
ax[1].imshow(im_data, cmap='gray', vmin=0, vmax=255)
ax[0].imshow(im_data, cmap='gray', vmin=0, vmax=255)

# Hide spines, ticks, etc.
ax.axis('off')

# Display the image.
ax.imshow(im_data, interpolation='nearest')
```

```python
import matplotlib.pyplot as plt
import cv2

imageA = plt.imread('test_images/solidWhiteRight.jpg')
imageB = plt.imread('test_images/solidWhiteRight.jpg')


# show the output images
print(type(imageA))
print(type(imageB))
cv2.imshow('imageA', imageA)
cv2.imshow('imageB', imageB)

while True:
    k = cv2.waitKey(30) & 0xFF
    if k == 27: # wait for ESC key to exit
        cv2.destroyAllWindows()
        cv2.waitKey(1)
        break
```
